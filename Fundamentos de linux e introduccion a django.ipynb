## Estructura de un Proyecto en Django

## Archivos de Proyecto

- manage.py es un arcivo que contiene funciones que permiten interactuar con Django
- config/__init__.py: un archivo vacio que le informa a Python que este directorio deberia ser consiedera como un paquete de Python
- config/settins.py: Configuracion o ajustes para el proyecto en django
- config/urls.py las declaraciones de los urls para el proyecto de django. es un gestor de urls
- config/wsgi.py un punto de entrada para servidores web compatibles con WSGI para desplegar proyecto

## Archivos de Aplicacion

- admin.py registrar tus modelos relacionados con el django admin
- apps.py contiene el codigo de la configuracion de la app
- models.py define tus modelos. son tablas en la base de datos para almacenar la infomacion y definir sus relaciones
- test.py: escribir test para la app
- views.py define la informacion que sera mandada a los templates de frontend

import os

from django import get_version
from django.conf import settings
from django.shortcuts import render


def home(request):
    context = {
        "debug": settings.DEBUG,
        "django_ver": get_version() + "Probando",
        "python_ver": os.environ["PYTHON_VERSION"] + "Texto agregado por Cordero",
    }

    return render(request, "pages/home.html", context)
